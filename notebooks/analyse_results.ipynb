```python
import pandas as pd
import datetime
import matplotlib.pyplot as plt
```

```python
df10 =pd.read_csv('../data/simulation_10_year.csv')
df15 = pd.read_csv('../data/15_year_investment_simulation.csv')
```

```python
df10['Return (%)'].hist()
```

```python
df15['Return (%)'].hist()
```

```python
df10['positive_return'] = df10['Return (%)'].apply(lambda x: 1 if x>=0 else 0)
df15['positive_return'] = df15['Return (%)'].apply(lambda x: 1 if x>=0 else 0)
```

```python
df10['positive_return'].value_counts(normalize=True)
```

```python
df15['positive_return'].value_counts(normalize=True)
```

```python
df10['Return (%)'].plot()
```

```python
df = df10

print(df['positive_return'].value_counts(normalize=True))

# Convert to datetime
df['First Investment Day'] = pd.to_datetime(df['First Investment Day'])
df['Last Investment Day'] = pd.to_datetime(df['Last Investment Day'])

df['First investment year'] = df['First Investment Day'].dt.year

# Calculate the midpoint year
df['Midpoint Year'] = df['First Investment Day'] + (df['Last Investment Day'] - df['First Investment Day']) / 2
df['Midpoint Year'] = df['Midpoint Year'].dt.year

df['average_return'] = df['Return (%)'].mean()
df['profitable'] = 0

# Plot the data
plt.figure(figsize=(10, 6))
plt.plot(df['First investment year'], df['Return (%)'], marker='o', label='Return (%)')
plt.plot(df['First investment year'], df['average_return'], label ='Average return', linestyle='--')
plt.plot(df['First investment year'], df['profitable'], label='Profitable', linestyle='--')

# Add labels and title
plt.xlabel('Start of investing - Year')
plt.ylabel('Return (%)')
plt.title('Investment Returns Over Time - 10 year')
plt.axhline(0, color='gray', linestyle='--', linewidth=1)  # Line at 0% for reference
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()

# Show the plot
plt.show()
```
